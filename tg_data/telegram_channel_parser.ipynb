```python
import requests
from bs4 import BeautifulSoup
import urllib.request
import selenium.webdriver as webdriver
import pandas as pd
```

```python
def parse(url):
    
    browser = webdriver.Chrome(executable_path='/usr/local/bin/chromedriver')
    browser.get(url)
    source = browser.page_source
    soup = BeautifulSoup(source, "html.parser")
    browser.close()

#     r = requests.get(url)
#     soup = BeautifulSoup(r.content, "lxml")


    # parse msg text
#     msgs = soup.find_all(True, {"class":["tgme_widget_message_text js-message_text before_footer", "tgme_widget_message_text js-message_text", "tgme_widget_message_photo"]})
#     msgs1 = soup.find_all('div', class_="tgme_widget_message_text js-message_text before_footer")
#     msgs2 = soup.find_all('div', class_="tgme_widget_message_text js-message_text")
#     msgs = msgs1 + msgs2

#     msg_list = []
#     for msg in msgs:
#         if msg.text is not '':
#             msg_list.append(msg.text)
#         else:
#             msg_list.append('jpg')
    
    # parse msg views count
    views = soup.find_all('span', class_="tgme_widget_message_views")
    
    view_list = []
    for view in views:
        view_list.append(view.text)
        
    # parse msg date and time
    date_times = soup.find_all('a', class_= "tgme_widget_message_date")
    
    date_list = []
    time_list = []
    dt_list = []
    for elem in date_times:
        raw = elem.time
        date = raw['datetime'][:10]
        date_list.append(date)
        time = raw.text
        time_list.append(time)

    # put all lists into 2D lists
#     content = list(zip(msg_list, view_list, date_list, time_list))
    
    return view_list, date_list, time_list
```

```python
def parse_all_pages():
    
    url1 = 'https://t.me/s/antielabfactchecking?before=26'
    view_list, date_list, time_list = parse(url1)

    for i in range(46, 680, 10):
        url = 'https://t.me/s/antielabfactchecking?before=' + str(i)
        x, y, z = parse(url)
#         msg_list.extend(a)
        view_list.extend(x)
        date_list.extend(y)
        time_list.extend(z)      

    content = list(zip(view_list, date_list, time_list))
    return content
```

```python
content = parse_all_pages()
df = pd.DataFrame(content, columns = ['Views', 'Date', 'Time'], index = False)
```

```python
df.drop_duplicates()
df['Views'] = df['Views'].apply(lambda x: str(x)[:-1])
```
